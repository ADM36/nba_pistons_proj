```python
L = []

team_df = pd.DataFrame()
for player in df['PLAYER_ID']:
    player_name = df['PLAYER'][df['PLAYER_ID'] == player].values[0]
    career = playercareerstats.PlayerCareerStats(player_id = player)
    df_player = career.get_data_frames()[0]
    df_year_player = df_player.loc[(df_player['SEASON_ID'].str[-2:] == YEAR[-2:]) &
                              (df_player['TEAM_ID'] == 1610612765)]
    df_year_player['NAME'] = player_name
    L.append(df_year_player)
```

```python
final_df = pd.concat(L, ignore_index=True)
```
